## Objective 1 :Prepare & QA the Data ##

```python
worldBank_df = pd.read_excel("../worldBank/WorldBank.xlsx") #Import the World Bank Excel file.

```

```python
worldBank_df.head(5)
```

```python
worldBank_df.info()
# Multiple missing values From "Birth rate,Crude and other following fields"
```

```python
# Changing Datatype - Noticed memory usage reduction
worldBank_df=worldBank_df.astype(
    {
    'Country Name':'string',
    'Country Code': 'string',
    'Region' : 'string',
    'IncomeGroup' : 'string',
     'Birth rate, crude (per 1,000 people)' :'float32',
    'Death rate, crude (per 1,000 people)' :'float32',
    'Electric power consumption (kWh per capita)' :'float32',
    'GDP (USD)' :'float32',
    'GDP per capita (USD)' :'float32',
    'Individuals using the Internet (% of population)' :'float32',
    'Life expectancy at birth (years)' :'float32',
    'Population density (people per sq. km of land area)' :'float32',
    'Unemployment (% of total labor force) (modeled ILO estimate)' :'float32',
    }                     
)
worldBank_df.info()
```

```python
worldBank_df.describe(include='all')
worldBank_df
```

```python
#worldBank_df['GDP (USD)'].fillna(worldBank_df['GDP (USD)'].mean(), inplace=True) # GDP has around 3K missing values. Replacing Missing Values with Overall Mean
worldBank_df
```

```python
# Use "GDP" and "GDP per Capita" to calculate population in millions and create a "Population (M)" column.
#Two more columns GDP (USD)(M) and  Population (M) in Data set
worldBank_df['GDP (USD)(M)']=(worldBank_df['GDP (USD)'])/1000000
worldBank_df['Population (M)'] = (worldBank_df['GDP (USD)(M)'] / worldBank_df['GDP per capita (USD)']).round(2)
#worldBank_df['Population (M)'].fillna(worldBank_df['Population (M)'].mean() ,inplace=True )
# changing Datatyp for two columns
worldBank_df=worldBank_df.astype(
    {
        'GDP (USD)(M)' : 'float32',
        'Population (M)' :'float32'
    }
)
```

```python
#Filter the data to 2014 
worldBank2014_df=worldBank_df.loc[worldBank_df['Year']==2014]
worldBank2014_df
```

```python
HDI_df = pd.read_csv("../worldBank/HDI.csv") # Reading Human Development Index Data
HDI_df_SelectedCols=['iso3','country','hdicode','region','hdi_2014']
```

```python
HDI_df=HDI_df[[(eachCol)for eachCol in HDI_df_SelectedCols ]]
HDI_df.head(100)
```

```python
HDI_df.tail(5)
```

```python
HDI_df.info()
```

```python
HDI_df.describe()
```

```python
#join it with the HDI csv file on "Country Code".
worldBank2014_HDI_mergedf=worldBank2014_df.merge(
        HDI_df,
        how='left',
        left_on=['Country Code'],
        right_on=['iso3']
    )

worldBank2014_HDI_mergedf
```

```python
worldBank2014_HDI_mergedf.info()
```

```python
worldBank_HDI_mergedf=worldBank_df.merge(
        HDI_df,
        how='left',
        left_on='Country Code',
        right_on=['iso3']
    )

worldBank_HDI_mergedf
```

## Objective 2 Prepare the data for visualization

```python
#Create a "gdp_pivot" table with years as rows, regions as columns, sum of GDP as values.
gdp_pivot2014=worldBank2014_HDI_mergedf.pivot_table(
    index='Year',
    columns='Region',
    values='GDP (USD)',
    aggfunc='sum'
)

gdp_pivot2014
```

```python
#Create a "gdp_pivot" table with years as rows, regions as columns, sum of GDP as values.
gdp_pivot=worldBank_df.pivot_table(
    index='Year',
    columns='Region',
    values='GDP (USD)(M)',
    aggfunc='sum'
)
gdp_pivot.head()
```

```python
#Create a "gdp_pop" table with years as rows, regions as columns, sum of population as values.
gdp_pop=worldBank_HDI_mergedf.pivot_table(
    index='Year',
    columns='Region',
    values='Population (M)',
    aggfunc='sum'
)
gdp_pop
```

```python
#Create "wb_hdi_by_region" table by calculating the average HDI for each region and sorting from highest average HDI to lowest.
worldBank_HDI14_mergedf_grbRegion=worldBank_HDI_mergedf.groupby('Region').agg(hdi2014_mean=('hdi_2014','mean')).sort_values('hdi2014_mean',ascending=[True])
worldBank_HDI14_mergedf_grbRegion
```

## Objective 3 :Visualize the data


```python
hexColorList=list(sns.color_palette("Set2").as_hex())
```

```python
# Map regions to colors
region_colors = {
    "East Asia & Pacific": hexColorList[0],
    "Europe & Central Asia": hexColorList[1],
    "Latin America & Caribbean": hexColorList[2],
    "Middle East & North Africa": hexColorList[3],
    "North America": hexColorList[4],
    "South Asia": hexColorList[5],
    "Sub-Saharan Africa": hexColorList[6]
}
stack1_List= [region_colors[region] for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index]
print (stack1_List)

stack2_List= [region_colors[region] for region in gdp_pop.iloc[-1].sort_values(ascending=False).index]
print (stack2_List)

bubble_List= stack2_List
```

```python
# Create a stacked area chart showing the growth of GDP over time, with each stack representing a geographic region.
fig, ax = plt.subplots(figsize=(9, 9))
ax.stackplot(
    gdp_pivot.index, 
    [gdp_pivot[region]/1000000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index], # sorting and Converting GDP in Trillion
    #gdp_pivot["East Asia & Pacific"]/1000000, # Converting GDP Millions to Trillions
    #gdp_pivot["Europe & Central Asia"]/1000000, 
    #gdp_pivot["Latin America & Caribbean"]/1000000,
    #gdp_pivot["Middle East & North Africa"]/1000000,
    #gdp_pivot["North America"]/1000000,
    #gdp_pivot["South Asia"]/1000000,
    #gdp_pivot["Sub-Saharan Africa"]/1000000,
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack1_List
    )

ax.set_title("Region's GDP Trend from 1960 to 2018")
ax.set_ylabel("GDP in trillion USD")
ax.legend(bbox_to_anchor=(1.4, 1));
ax.set_xlim(1960,2020);

```

```python
#Create a stacked area chart showing the growth of population over time, with each stack representing a geographic region.
fig, ax = plt.subplots(figsize=(9, 9))
ax.stackplot(
    gdp_pop.index, 
    [gdp_pop[region] for region in gdp_pop.iloc[-1].sort_values(ascending=False).index],
    #gdp_pop["East Asia & Pacific"], 
    #gdp_pop["Europe & Central Asia"], 
    #gdp_pop["Latin America & Caribbean"],
    #gdp_pop["Middle East & North Africa"],
    #gdp_pop["North America"],
    #gdp_pop["South Asia"],
    #gdp_pop["Sub-Saharan Africa"],
    labels=gdp_pop.iloc[-1].sort_values(ascending=False).index,
    colors=stack2_List
    )

ax.set_title("Region's Population Trend from 2009 to 2018")
ax.set_ylabel("Population(M)")
ax.legend(bbox_to_anchor=(1.4, 1));
ax.set_xlim(1960,2020);
```

```python
#Create a bar chart to show average HDI by region (make sure the region colors match).
fig, ax = plt.subplots(figsize=(9, 9))
ax.set_title("HDI 2014 Index")
ax.set_xlabel("Region")
colors = [region_colors[region] for region in worldBank_HDI14_mergedf_grbRegion.index]
ax.bar(worldBank_HDI14_mergedf_grbRegion.index, worldBank_HDI14_mergedf_grbRegion['hdi2014_mean'], color=colors)
plt.xticks(rotation=45)
plt.show()
```

```python
import  matplotlib.gridspec as gridspec
fig=plt.figure(figsize=(14,14))
gs=gridspec.GridSpec(nrows=13,ncols=12)

fig.suptitle("Report : Global Economic Development from 1960-2018",fontsize=16,fontweight='bold',x=0.50,y=0.93)
            

ax1=fig.add_subplot(gs[0:4,0:4])
ax1.stackplot(
    gdp_pivot.index, 
    [gdp_pivot[region]/1000000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index], # sorting and Converting GDP in Trillion
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack1_List
    )

ax1.set_title("Region's GDP Trend from 1960 to 2018")
ax1.set_ylabel("GDP in trillion USD")

#ax1.set_xlim(1960,2020);


ax2=fig.add_subplot(gs[0:4,5:9])
ax2.stackplot(
    gdp_pop.index, 
    [gdp_pop[region]/1000 for region in gdp_pop.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pop.iloc[-1].sort_values(ascending=False).index,
    colors=stack2_List
    )

ax2.set_title("Region's Population Trend from 1960 to 2018")
ax2.set_ylabel("Population(Billion)")
ax1.legend(bbox_to_anchor=(3.0, 1),fontsize=9);
fig.text(0.72,0.65,'''
From 1960 to 2018, the region's GDP surged from 0 trillion USD to 80 trillion USD. 
Approximately 50% of this GDP is contributed by 
'East Asia & Pacific' and 'Europe & Central Asia'.
During the same period, the population grew from 2 billion to 7 billion. 
The highest population growth was observed in the 
'East Asia and Pacific' region. Overall, the increase in GDP is
largely attributed to the increase in population.
''')

ax3=fig.add_subplot(gs[5:9,0:9])
minsize=min(worldBank_HDI_mergedf.query("Year==2014")['Population (M)'])
maxsize=max(worldBank_HDI_mergedf.query("Year==2014")['Population (M)'])
sns.scatterplot(
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes=(minsize,maxsize),
    data=worldBank_HDI_mergedf.query("Year==2014"),
    hue=worldBank_HDI_mergedf["Region"],
    palette=stack2_List,
    ax=ax3
   ).set(title='Life Expectancy Increases as Countries get richer',yscale='log');

handles,labels=ax3.get_legend_handles_labels()
entries_to_skip=len(worldBank_HDI_mergedf['Region'].unique())+1

for h in handles[1:]:
    sizes=[s/1.5 for s in h.get_sizes()]
    h.set_sizes(sizes)
    
ax3.legend(
    handles[entries_to_skip:],
    labels[entries_to_skip:],
    bbox_to_anchor=(.8,.9),
    borderaxespad=0,
    frameon=False,
    ncol=6,
    fontsize=8
);


ax4=fig.add_subplot(gs[10:13,0:4])
ax4.set_title("HDI 2014 Index")
#ax4.set_xlabel("Region")
colors = [region_colors[region] for region in worldBank_HDI14_mergedf_grbRegion.index]
ax4.bar(worldBank_HDI14_mergedf_grbRegion.index, worldBank_HDI14_mergedf_grbRegion['hdi2014_mean'], color=colors)
plt.xticks(rotation=90)



ax5=fig.add_subplot(gs[10:13,5:9])
sns.scatterplot(
    data=worldBank_HDI_mergedf.query("Year==2014 and 'Country Name'  !='Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette='coolwarm_r',
    ax=ax5
   ).legend(bbox_to_anchor=(1, 1));

fig.text(0.72,0.09,'''
In 2014, North America had the highest Human Development Index (HDI). 
There is a strong correlation between GDP per capita and electric power consumption. 
The highest electric consumption is typically observed in regions with a 
high Human Development Index.''');


```
